# Interpolação de variáveis

## Método format

```python
nome = "Bruno"
idade = 39
profissao = "Desenvolvedor"
linguagem = "Python"

pessoa = {
  "nome": "Bruno",
  "idade": 39,
  "profissao": "Desenvolvedor",
  "linguagem": "Python",
}
```

```python
"Olá, me chamo {}. Eu tenho {} anos, trabalho como {} e estou estudando a linguagem {}".format(nome, idade, profissao, linguagem)
```

```python
"Olá, me chamo {3}. Eu tenho {2} anos, trabalho como {1} e estou estudando a linguagem {0}".format(linguagem, profissao, idade, nome)
```

```python
"Olá, me chamo {nome}. Eu tenho {idade} anos, trabalho como {profissao} e estou estudando a linguagem {linguagem}".format(nome=nome, idade=idade, profissao=profissao, linguagem=linguagem)
```

```python
"Olá, me chamo {nome}. Eu tenho {idade} anos, trabalho como {profissao} e estou estudando a linguagem {linguagem}".format(**pessoa)
```

## f-string

```python
nome = "Bruno"
idade = 39
profissao = "Desenvolvedor"
linguagem = "Python"
```

```python
f"Olá, me chamo {nome}. Eu tenho {idade} anos, trabalho como {profissao} e estou estudando a linguagem {linguagem}"
```

### Formatar string com f-string

```python
PI = 3.14159
```

```python
f"Valor de PI: {PI:.2f}"
```

```python
f"Valor de PI: {PI:10.3f}"
```
