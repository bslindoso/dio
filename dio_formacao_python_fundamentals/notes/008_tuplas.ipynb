# Tuplas

Tuplas são estruturas de dados em Python que permitem armazenar uma sequência organizada de elementos de forma imutável. Diferente das listas, uma vez que uma tupla é criada, seus valores não podem ser alterados, adicionados ou removidos, o que garante a integridade dos dados ao longo da execução do programa. Elas são ideais para agrupar informações relacionadas que devem permanecer constantes, oferecendo maior segurança e eficiência no processamento de coleções fixas.

```python
frutas = ("maçã", "banana", "laranja",)

letras = tuple("python")

numeros = tuple([1, 2, 3, 4, 5])

pais = ("Brasil",)
```

```python
frutas[0]
```

## Tuplas aninhadas

```python
matriz = (
  (1, "a", 3), 
  ("b", 5, 6), 
  (7, 8, "c")
)
```

```python
print(matriz[0])
print(matriz[0][0])
print(matriz[0][-1])
print(matriz[-1][-1])
```

## Fatiamento

```python
lista = ("p", "y", "t", "h", "o", "n")
```

```python
print(lista[2:])
print(lista[:2])
print(lista[0:3:2])
print(lista[::])
print(lista[::-1])
```
